# Converting arff to csv using arff

#### The original dataset from <http://archive.ics.uci.edu/dataset/545/rice+cammeo+and+osmancik> is in arff I will convert to csv which is easier to load

```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('rice+cammeo+and+osmancik/Rice_Cammeo_Osmancik.arff')
data
```

```python
df = pd.DataFrame(data[0])
```

```python
df.to_csv('Rice_Classification.csv', index=False)
```
